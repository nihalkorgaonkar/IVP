```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Read the colored image
img_color = cv2.imread('Nihalimg2.jpg')

# Step 2: Convert the image to grayscale
img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

# Step 3: Display the grayscale image
plt.figure(figsize=(6, 6))
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image - Nihalimg2")
plt.axis('off')
plt.show()

# Step 4: Plot histogram
hist = cv2.calcHist([img_gray], [0], None, [256], [0, 256])
plt.figure(figsize=(8, 4))
plt.plot(hist, color='black')
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Intensity (0-255)')
plt.ylabel('Number of Pixels')
plt.grid(True)
plt.show()

# Step 5: Basic statistical analysis
print("---- Histogram Analysis ----")
print(f"Minimum Pixel Value     : {img_gray.min()}")
print(f"Maximum Pixel Value     : {img_gray.max()}")
print(f"Mean Intensity Value    : {img_gray.mean():.2f}")
print(f"Standard Deviation      : {img_gray.std():.2f}")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Read the image and convert to grayscale
img_color = cv2.imread('Nihalimg2.jpg')
img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

# Step 2: Apply Histogram Equalization
img_eq = cv2.equalizeHist(img_gray)

# Step 3: Compute Histograms
hist_orig = cv2.calcHist([img_gray], [0], None, [256], [0, 256])
hist_eq = cv2.calcHist([img_eq], [0], None, [256], [0, 256])

# Step 4: Display results
plt.figure(figsize=(14, 8))

# Original Grayscale Image
plt.subplot(2, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

# Histogram of Original Image
plt.subplot(2, 2, 2)
plt.plot(hist_orig, color='black')
plt.title('Histogram - Original Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Number of Pixels')
plt.grid(True)

# Equalized Image
plt.subplot(2, 2, 3)
plt.imshow(img_eq, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')

# Histogram of Equalized Image
plt.subplot(2, 2, 4)
plt.plot(hist_eq, color='black')
plt.title('Histogram - Equalized Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Number of Pixels')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read the color image
img_color = cv2.imread('Nihalimg2.jpg')

# Split the image into B, G, R channels
colors = ('b', 'g', 'r')
plt.figure(figsize=(8, 4))
for i, col in enumerate(colors):
    hist = cv2.calcHist([img_color], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])

plt.title('Histogram of Color Image (BGR Channels)')
plt.xlabel('Pixel Intensity')
plt.ylabel('Number of Pixels')
plt.grid(True)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Step 1: Read the original color image
img_color = cv2.imread('Nihalimg2.jpg')

# Step 2: Convert image to YCrCb color space
img_ycrcb = cv2.cvtColor(img_color, cv2.COLOR_BGR2YCrCb)

# Step 3: Split channels
y, cr, cb = cv2.split(img_ycrcb)

# Step 4: Apply histogram equalization to the Y channel only
y_eq = cv2.equalizeHist(y)

# Step 5: Merge equalized Y with original Cr and Cb
img_eq_ycrcb = cv2.merge([y_eq, cr, cb])

# Step 6: Convert back to BGR color space
img_eq_color = cv2.cvtColor(img_eq_ycrcb, cv2.COLOR_YCrCb2BGR)

# Step 7: Display original and equalized color images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title('Original Color Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_eq_color, cv2.COLOR_BGR2RGB))
plt.title('Histogram Equalized Color Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
